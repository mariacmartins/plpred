# Exploratory Data Analysis
O **plpred** é uma aplicação de **predição de localização subcelular de proteínas** a partir da sua composição de aminoácidos. 

Esse notebook refere-se à análise exploratória dos dados da aplicação **plpred**. De modo geral, a análise exploratória de dados é uma abordagem à análise de conjuntos de dados de modo a resumir suas características principais, frequentemente com métodos visuais.

## Data Preprocessing

```python
from Bio import SeqIO
from Bio.SeqUtils import ProtParam
import pandas as pd
```

```python
def compute_aa_composition(protein_sequence):
    analyzer = ProtParam.ProteinAnalysis(str(protein_sequence))
    return analyzer.get_amino_acids_percent()
```

```python
#criando um campo “membrane” no df: 1 para as proteínas de membrana e 0 para as proteínas citoplasmáticas. 
df = pd.DataFrame()

handle = open('../data/raw/membrane.fasta')
parser = SeqIO.parse(handle, 'fasta')

for protein in parser:
    protein_data = compute_aa_composition(protein.seq)
    protein_data['membrane'] = 1
    df = df.append([protein_data], ignore_index=True)
    
handle = open('../data/raw/cytoplasm.fasta')
parser = SeqIO.parse(handle, 'fasta')

for protein in parser:
    protein_data = compute_aa_composition(protein.seq)
    protein_data['membrane'] = 0
    df = df.append([protein_data], ignore_index=True)
```

## Exploratory Data Analysis

```python
df_grouped = df.groupby('membrane').mean()
df_grouped
```

```python
%pylab inline

pylab.rcParams['figure.figsize'] = (12, 6)
aminoacids = df_grouped.columns

cytoplasm_mean_composition = df_grouped.loc[0]
membrane_mean_composition = df_grouped.loc[1]

relative_difference = membrane_mean_composition / cytoplasm_mean_composition

plt.bar(aminoacids, relative_difference)
plt.show()
```

**Características físico-químicas:** Proteínas de membrana possuem características mais hidrofóbicas para interagir com ácidos graxos em relação à proteínas citoplasmáticas.

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca.fit(df.drop(['membrane'], axis=1))

for group, df_group in df.groupby('membrane'):
    components = pca.transform(df_group.drop('membrane', axis=1))
    plt.scatter(
        components[:,0],
        components[:,1],
        s=0.5,
        label='Membrane' if group == 0 else 'Cytoplasm',
        color='red' if group == 0 else 'blue'
    )

plt.show()
```

**Análise de Componente Principal (Principal Component Analysis, PCA)**: Análise para ver, em duas dimensões, a diferença entre as proteínas de membrana e as proteínas citoplasmáticas.

A partir das características de composição de aminoácidos foi possível observar que houve grupos de proteínas que por conta das suas características são quase que exclusivamente proteínas de membranas e outras proteínas que são quase que exclusivamente proteínas de citoplasma.
Esse resultado propõe um bom indicativo que o modelo conseguirá aprender a discriminar proteinas de membrana e citoplasma.

## Model Training

```python
from sklearn.model_selection import train_test_split

X = df.drop(['membrane'], axis=1)
y = df['membrane']

X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_train, y_train)
```

## Model Validation

```python
from sklearn.metrics import classification_report

y_pred = model.predict(X_test)
report = classification_report(y_test, y_pred)
print(report)
```

**Métricas:**
- Acurácia: indica uma performance geral do modelo. Dentre todas as classificações, quantas o modelo classificou corretamente.
- Precisão: dentre todas as classificações de classe positivo que o modelo fez, quantas estão corretas.
- Recall: dentre todas as situações de classe positivo como valor esperado, quantas estão corretas.
- F1-Score: média harmônica entre precisão e recall.

**Classificação:**
- Proteínas de membrana: 1
- Proteínas citoplasmáticas: 0

**Conclusão geral:** O modelo apresentou ótimos valores nas métricas de avaliação e consequentemente boa perfomance, com um bom valor de acurácia e f1-score.
